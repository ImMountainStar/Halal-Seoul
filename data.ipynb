```python
# RAWMTRL_NM 열만 사용
row_text = data['RAWMTRL_NM'].fillna('').astype(str)
```

```python
# 쉼표(,) 기준으로 분리 (괄호 내부 쉼표는 분리하지 않음)
def split_materials(text):
    s = str(text)
    tokens = []
    buf = []
    depth = 0
    for ch in s:
        if ch == '(':
            depth += 1
        elif ch == ')' and depth > 0:
            depth -= 1

        if ch == ',' and depth == 0:
            token = ''.join(buf).strip()
            if token:
                tokens.append(token)
            buf = []
        else:
            buf.append(ch)

    token = ''.join(buf).strip()
    if token:
        tokens.append(token)
    return tokens

data_list = row_text.apply(split_materials)
data_list
```

```python
# 모든 행의 값을 하나로 합치고 중복 제거
unique_values = sorted({item for row in data_list for item in row})

```

```python
unique_values
```

```python
# 고유 값 데이터프레임
unique_df = pd.DataFrame({'raw_name': unique_values})
#unique_df.head(50)
```

```python
# 고유 원재료명 정규화 (실데이터 패턴 반영)
import re
import unicodedata

''' 
GREEK_MAP = str.maketrans({
    'α': ' alpha ', 'β': ' beta ', 'γ': ' gamma ', 'δ': ' delta ',
    'ε': ' epsilon ', 'ω': ' omega ', '∝': ' alpha '
})
'''

ROMAN_MAP = str.maketrans({
    'Ⅰ': ' I ', 'Ⅱ': ' II ', 'Ⅲ': ' III ', 'Ⅳ': ' IV ',
    'Ⅴ': ' V ', 'Ⅵ': ' VI ', 'Ⅶ': ' VII ', 'Ⅷ': ' VIII ', 'Ⅸ': ' IX ', 'Ⅹ': ' X ',
    'ⅰ': ' i ', 'ⅱ': ' ii ', 'ⅲ': ' iii ', 'ⅳ': ' iv ', 'ⅴ': ' v '
})

PUNCT_FIX = str.maketrans({
    '’': "'", '‘': "'", '´': "'", '`': "'", '＇': "'",
    '₂': '2', '₃': '3', '₄': '4', '₅': '5', '₆': '6', '₇': '7', '₈': '8', '₉': '9', '₀': '0'
})

def normalize_material_name(x):
    s = '' if pd.isna(x) else str(x)
    s = unicodedata.normalize('NFKC', s)
    s = s.translate(PUNCT_FIX)
    #s = s.translate(GREEK_MAP)
    s = s.translate(ROMAN_MAP)
    s = s.lower()  # 영문 대소문자 통일

    # 괄호/특수 구분자 정리
    #s = re.sub(r'[\[\]\(\)\{\}]', ' ', s)
    #s = re.sub(r'[\/·ㆍ&+]', ' ', s)

    # 허용 문자만 남기기 (한글/영문/숫자/공백/하이픈/작은따옴표)
    s = re.sub(r"[^0-9a-z가-힣\s\-'()]", ' ', s)

    # 기호 주변 공백 정리
    s = re.sub(r'\s*-\s*', '-', s)
    s = re.sub(r"\s*'\s*", "'", s)

    # 연속 공백 1칸화 및 양끝 정리
    s = re.sub(r'\s+', ' ', s).strip(" -'")
    return s


def is_valid_material_token(s):
    # 길이 1 이하 제거
    if len(s) <= 1:
        return False
    # 숫자/기호만 구성된 토큰 제거 (예: 1, 2, 95, --)
    if re.fullmatch(r"[0-9\s\-'.]+", s):
        return False
    # 최소 1개 이상의 한글/영문이 있어야 함
    if not re.search(r"[a-z가-힣]", s):
        return False
    return True

unique_df['normalized_name'] = unique_df['raw_name'].apply(normalize_material_name)
unique_df = unique_df[unique_df['normalized_name'].apply(is_valid_material_token)]
unique_df = unique_df.drop_duplicates('normalized_name').reset_index(drop=True)

```

```python
# 공백 제거 
unique_df['normalized_name'] = (
    unique_df['normalized_name']
    .fillna('')
    .astype(str)
    .str.replace(r'\s+', '', regex=True)
)

# 빈값 제거 + canonical_name 기준 중복 제거
unique_df = (
    unique_df[unique_df['normalized_name'] != '']
    .drop_duplicates(subset='normalized_name')
    .reset_index(drop=True)
)
```
